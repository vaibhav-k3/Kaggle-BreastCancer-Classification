```python
X_data = data[['area_mean','concave points_mean','texture_mean','compactness_mean','concavity_mean']]
#X_data.loc[:,'area_mean'] = (X_data.loc[:,'area_mean'] - np.mean(X_data['area_mean']))/np.max(X_data['area_mean'])
#X_data.loc[:,'concave points_mean'] = (X_data.loc[:,'concave points_mean'] - np.mean(X_data['concave points_mean']))/np.max(X_data['concave points_mean'])
#X_data.loc[:,'texture_mean'] = (X_data.loc[:,'texture_mean'] - np.mean(X_data['texture_mean']))/np.max(X_data['texture_mean'])
#X_data.loc[:,'compactness_mean'] = (X_data.loc[:,'compactness_mean'] - np.mean(X_data['compactness_mean']))/np.max(X_data['compactness_mean'])
#X_data.loc[:,'concavity_mean'] = (X_data.loc[:,'concavity_mean'] - np.mean(X_data['concavity_mean']))/np.max(X_data['concavity_mean'])
Y_data = pd.DataFrame(index= data.index , columns = ['M'])
for col in X_data.columns:
    X_data.loc[:,col] = (X_data.loc[:,col] - np.mean(X_data[col]))/np.max(X_data[col])
    
for i in Y_data.index:
    if data.loc[i , 'diagnosis'] == 'M':
        Y_data.loc[i,'M'] = 1
    else:
        Y_data.loc[i , 'M'] = 0
        
X_data.head()

#not able to resolve this warning
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X_data , Y_data , test_size = 0.2)
```

```python

for k in range(1,15):
    Knnmodel = KNeighborsClassifier(n_neighbors = k , weights = 'uniform')
    Knnmodel.fit(np.array(X_train) , np.array(Y_train).ravel()) ## the ravel function flattens the array
    #what ravel() does is: when you have y.shape == (10, 1), using y.ravel().shape == (10, ). In words... it flattens an array.
    #x=Knnmodel.score(X_test , Y_test)
    x = 0
    for i in Y_test.index:
        if Knnmodel.predict(np.array(X_test.loc[i]).reshape(1,-1))== Y_test.loc[i , 'M']:
            x = x +1
    x = (x/len(Y_test))
    print('k = '+str(k) + '-->' + str(x))
    
          
```
